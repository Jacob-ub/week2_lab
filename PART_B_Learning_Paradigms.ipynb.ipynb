```python
import numpy as np
from sklearn.linear_model import LinearRegression

# modify areas and prices
areas = np.array([60, 85, 100, 125, 150])
prices = np.array([30000, 42000, 50000, 63000, 75000])

# train model
X = areas.reshape(-1, 1)
y = prices

model = LinearRegression()
model.fit(X, y)

# predictions
print("Prediction for 80 m2:", model.predict([[80]])[0])
print("Prediction for 140 m2:", model.predict([[140]])[0])
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# extend X and y
X = np.array([
    [80, 20],
    [90, 18],
    [60, 25],
    [30, 30],
    [85, 19],
    [70, 23],   # new sample
    [95, 16]    # new sample
])

y = np.array([1, 1, 0, 0, 1, 1, 1])

# train classifier
clf = LogisticRegression()
clf.fit(X, y)

# predictions
print("Prediction for [60,24]:", clf.predict([[60, 24]])[0])
print("Prediction for [90,17]:", clf.predict([[90, 17]])[0])
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# modify data (added more houses)
data = np.array([
    [50, 25000],
    [60, 30000],
    [80, 40000],
    [100, 50000],
    [120, 60000],
    [140, 70000],
    [160, 80000]
])

# clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=clusters)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Clusters (3 Groups)")
plt.show()
```

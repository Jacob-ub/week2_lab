```python
areas = [55, 75, 95, 115, 135]
prices = [27000, 38000, 47000, 58000, 69000]

print(areas)
print(prices)
```

```python
import numpy as np

areas_np = np.array(areas)
prices_np = np.array(prices)

("Average price:", np.mean(prices_np))
print("Max price:", np.max(prices_np))
print("Min area:", np.min(areas_np))
```

```python
import matplotlib.pyplot as plt

plt.scatter(areas_np, prices_np)
plt.xlabel("House Area (m2)")
plt.ylabel("House Price ($)")
plt.title("My House Price vs Area Data")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression


areas_np = np.array([60, 80, 100, 120, 140])
prices_np = np.array([30000, 42000, 52000, 63000, 75000])


X = areas_np.reshape(-1, 1)
y = prices_np

model = LinearRegression()
model.fit(X, y)

print("Prediction for 70 m2:", model.predict([[70]])[0])
print("Prediction for 110 m2:", model.predict([[110]])[0])
print("Prediction for 150 m2:", model.predict([[150]])[0])
```
